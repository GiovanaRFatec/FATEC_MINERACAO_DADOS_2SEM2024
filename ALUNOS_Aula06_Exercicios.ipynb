# 01 - Criação e Manipulação de Arrays

Crie um array Numpy com 10 números inteiros aleatórios entre 1 e 20 e extraia os elementos pares.

```python
import numpy as np
# Criação do array com 10 números inteiros aleatórios entre 1 e 20
array = np.random.randint(1, 21, size=10)

# Extração dos elementos pares
pares = array[array % 2 == 0]

print("Array original:", array)
print("Elementos pares:", pares)
```

# 02 - Operações Aritméticas com Arrays

Crie dois arrays de 10 números inteiros aleatórios entre 0 e 50 e realize operações de soma, subtração, multiplicação e divisão elemento a elemento.

```python
import numpy as np

# Criação dos dois arrays com 10 números inteiros aleatórios entre 0 e 50
array1 = np.random.randint(0, 51, size=10)
array2 = np.random.randint(0, 51, size=10)

# Operações elemento a elemento
soma = array1 + array2
subtracao = array1 - array2
multiplicacao = array1 * array2
divisao = np.divide(array1, array2, where=array2!=0)  # Evitar divisão por zero

print("Array 1:", array1)
print("Array 2:", array2)
print("Soma:", soma)
print("Subtração:", subtracao)
print("Multiplicação:", multiplicacao)
print("Divisão:", divisao)
```

# 03 - Indexação e Slicing

Dado um array Numpy com 20 números aleatórios entre 10 e 100, extraia os elementos entre a 5ª e a 15ª posição.

```python
import numpy as np

# Criação do array com 20 números inteiros aleatórios entre 10 e 100
array = np.random.randint(10, 101, size=20)

# Extração dos elementos entre a 5ª e a 15ª posição
subarray = array[4:15]  # O índice começa em 0, então 5ª posição é o índice 4, e 15ª é o índice 14

print("Array original:", array)
print("Elementos entre a 5ª e a 15ª posição:", subarray)
```

# 04 - Manipulação de Matrizes

Crie uma matriz 3x3 com valores aleatórios entre 0 e 1, e encontre o valor máximo e mínimo.

```python
import numpy as np

# Criação da matriz 3x3 com valores aleatórios entre 0 e 1
matriz = np.random.rand(3, 3)

# Encontrar o valor máximo e mínimo
valor_maximo = np.max(matriz)
valor_minimo = np.min(matriz)

print("Matriz 3x3:")
print(matriz)
print("Valor máximo:", valor_maximo)
print("Valor mínimo:", valor_minimo)
```

# 08 - Filtragem de Dados

Filtre o DataFrame criado no exercício anterior para mostrar apenas as vendas onde a quantidade é maior que 10.

```python
import pandas as pd

# Criando o DataFrame com os dados fornecidos
data = {
    'Produto': ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E'],
    'Quantidade': [5, 1, 17, 9, 1],
    'Preço': [55, 100, 16, 25, 84]
}
df = pd.DataFrame(data)

# Adicionando a coluna de valor total da venda
df['Valor Total'] = df['Quantidade'] * df['Preço']

# Filtrando o DataFrame para mostrar apenas as vendas onde a quantidade é maior que 10
df_filtrado = df[df['Quantidade'] > 10]

# Exibindo o DataFrame filtrado com a nova coluna
print(df_filtrado)
```

# 17 - Gráfico de Dispersão com Seaborn

Recrie o gráfico de dispersão da relação entre altura e peso usando Seaborn, incluindo uma linha de regressão.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Gerando dados fictícios para altura e peso de 100 pessoas
np.random.seed(0)  # Para reprodutibilidade
altura = np.random.normal(loc=170, scale=10, size=100)  # Altura em cm
peso = np.random.normal(loc=70, scale=15, size=100)  # Peso em kg

# Criando um DataFrame
df = pd.DataFrame({'Altura': altura, 'Peso': peso})

# Criando o gráfico de dispersão com linha de regressão
plt.figure(figsize=(5, 3))
sns.regplot(x='Altura', y='Peso', data=df, scatter_kws={'color': 'coral', 's': 100, 'edgecolor': 'black'}, line_kws={'color': 'blue'})

# Adicionando título e rótulos
plt.title('Relação entre Altura e Peso com Linha de Regressão')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')

# Exibindo o gráfico
plt.show()

```

# 19 - Heatmap

Crie um heatmap para mostrar a correlação entre diferentes variáveis em um DataFrame fictício.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Gerando dados fictícios
np.random.seed(0)  # Para reprodutibilidade
data = {
    'Variável A': np.random.normal(loc=50, scale=10, size=100),
    'Variável B': np.random.normal(loc=60, scale=15, size=100),
    'Variável C': np.random.normal(loc=70, scale=20, size=100),
    'Variável D': np.random.normal(loc=80, scale=25, size=100)
}

# Criando um DataFrame
df = pd.DataFrame(data)

# Calculando a matriz de correlação
correlation_matrix = df.corr()

plt.figure(figsize=(6, 3))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1, linewidths=0.5)

# Adicionando título
plt.title('Matriz de Correlação')

# Exibindo o gráfico
plt.show()
```

# 20 - Pairplot

Crie um pairplot para explorar relações entre várias variáveis em um conjunto de dados fictício.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Gerando dados fictícios
np.random.seed(0)  # Para reprodutibilidade
data = {
    'Variável A': np.random.normal(loc=50, scale=10, size=100),
    'Variável B': np.random.normal(loc=60, scale=15, size=100),
    'Variável C': np.random.normal(loc=70, scale=20, size=100),
    'Variável D': np.random.normal(loc=80, scale=25, size=100)
}

# Criando um DataFrame
df = pd.DataFrame(data)

# Criando o pairplot
sns.pairplot(df, diag_kind='kde', plot_kws={'alpha': 0.6, 's': 50, 'edgecolor': 'w'}, diag_kws={'color': 'blue'})

# Adicionando título
plt.suptitle('Pairplot das Variáveis', y=1.02)  # Ajuste y para título

# Exibindo o gráfico
plt.show()
```
